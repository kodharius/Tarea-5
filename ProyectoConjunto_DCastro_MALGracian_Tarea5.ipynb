El método de la transformada inversa es una técnica que nos permite generar variables aleatorias que siguen una distribución de probabilidad específica a partir de una distribución uniforme en el intervalo [0,1].

Primero, necesitamos calcular la distribución acumulada de la función de densidad g(x). La distribución acumulada se define como:

$$G(x) = \int_{-\infty}^{x} g(t) dt$$

Para nuestra función de densidad g(x) = a sin(πx), la distribución acumulada se puede calcular de la siguiente manera:

$$G(x) = \int_{-\infty}^{x} g(t) dt = \int_{-\infty}^{x} a \sin(\pi t) dt = \frac{-a}{\pi} \cos(\pi x) + \frac{a}{\pi}$$

Ahora, necesitamos invertir la distribución acumulada G(x) para obtener la función inversa G⁻¹(u) que se usa para generar variables aleatorias. La función inversa se puede calcular de la siguiente manera:

$$u = G(x) = \frac{-a}{\pi} \cos(\pi x) + \frac{a}{\pi}$$

Despejando x, obtenemos la función inversa:

$$x = G^{-1}(u) = \frac{1}{\pi} \cos^{-1}(\frac{\pi u - 1}{a})$$

```python
# Ilustración del método de aceptación y rechazo cuando se toma t(x) constante

# Función objetivo 
f = lambda x: 30 * (x**2 - 2*x**3 + x**4)
# Máximo de la función f
max_f = f(optimize.fmin(lambda x:-f(x), 0, disp=False))
# Función t -> Función constante
t = lambda x: max_f * np.sin(np.pi * x)

# Rango donde se graficará las funciones
x = np.arange(0, 1, 0.01)
print('El máximo de f es:',max_f)

# Gráficas de las funciones
plt.plot(x,f(x),label='f(x)')
plt.plot(x,t(x),label='t(x)')
plt.legend()

# Validación del método
N = 500 # número de puntos a simular
# Como estoy tomando t(x) constante solo es necesario generar valores aleatorios U~(0,1)
R2 = np.random.rand(N)
R1 = np.random.rand(N)

# Prueba de función creada
Acep_rechazo(R2, R1, f, t)
```

Para demostrar la validez de la ecuación dada, podemos utilizar inducción matemática.

**Paso base:** Para i = 0, la ecuación se convierte en:
$$p_1 = \frac{n}{1}\frac{p}{1-p} p_0$$
Lo cual es equivalente a $p_1 = \frac{n}{1}\frac{p}{1-p} p_0$, que es verdadero.

**Paso de inducción:** Supongamos que la ecuación es válida para algún valor k, es decir:
$$p_{k+1} = \frac{n-k}{k+1}\frac{p}{1-p} p_k$$

**Paso inductivo:** Consideremos i = k + 1:
$$p_{k+2} = \frac{n-(k+1)}{(k+1)+1}\frac{p}{1-p} p_{k+1}$$
Esto se simplifica a:
$$p_{k+2} = \frac{n-k-1}{k+2}\frac{p}{1-p} \left(\frac{n-k}{k+1}\frac{p}{1-p} p_k\right)$$
$$p_{k+2} = \frac{n-k-1}{k+2}\frac{n-k}{k+1}\frac{p}{1-p} \frac{p}{1-p} p_k$$
$$p_{k+2} = \frac{(n-k)(n-k-1)}{(k+1)(k+2)}\frac{p^2}{(1-p)^2} p_k$$
Esto es exactamente la forma de la ecuación para i = k + 1, por lo tanto, se mantiene válida para este caso.

Por lo tanto, la ecuación es válida para todos los enteros no negativos i por inducción matemática.


# Pregunta 9 Estudiante 1

Para demostrar la relación $P(k+1) = \frac{\lambda}{k+1}P(k)$, podemos seguir los siguientes pasos:

\begin{align*}
\frac{P(k+1)}{P(k)} &= \frac{\frac{e^{-\lambda}\lambda^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}\lambda^k}{k!}} \\
&= \frac{e^{-\lambda}\lambda^{k+1}}{(k+1)!} \cdot \frac{k!}{e^{-\lambda}\lambda^k} \\
&= \frac{e^{-\lambda}\lambda^{k+1}}{e^{-\lambda}\lambda^k} \cdot \frac{k!}{(k+1)!} \\
&= \frac{\lambda}{k+1}
\end{align*}

Por lo tanto, queda demostrado que $P(k+1) = \frac{\lambda}{k+1}P(k)$.

Ahora, para generar variables aleatorias que sigan una distribución de Poisson utilizando el método de la transformada inversa, podemos usar la relación demostrada:

Dado un valor $\lambda > 0$, podemos generar una variable aleatoria Poisson $X$ de la siguiente manera:

1. Generamos una variable aleatoria $U$ uniformemente distribuida en el intervalo $(0,1)$.
2. Encontramos el valor de $k$ tal que $\sum_{i=0}^{k} P(i) \leq U < \sum_{i=0}^{k+1} P(i)$.
3. Devolvemos el valor $k$ como el valor de nuestra variable aleatoria $X$.
